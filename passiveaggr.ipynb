```python
!pip install vaderSentiment


from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Perform sentiment analysis on the statements and print the results
for statement in df['Statement']:
    sentiment_scores = analyzer.polarity_scores(statement)
    compound_score = sentiment_scores['compound']
    sentiment_label = 'Positive' if compound_score > 0 else ('Negative' if compound_score < 0 else 'Neutral')
    print(f"Statement: '{statement}'")
    print(f"Sentiment Label: {sentiment_label} (Compound Score: {compound_score:.2f})")
    print("-----------------------------")

# Example: Classify user input statement
user_input = "I appreciate your input."
user_sentiment = analyzer.polarity_scores(user_input)
user_compound_score = user_sentiment['compound']
user_sentiment_label = 'Positive' if user_compound_score > 0 else ('Negative' if user_compound_score < 0 else 'Neutral')

print(f"User input statement: '{user_input}'")
print(f"Sentiment Label: {user_sentiment_label} (Compound Score: {user_compound_score:.2f})")
```

###**Text preprocessing and feature extraction**

```python

# Text preprocessing and feature extraction
def preprocess_text(text):
    # Lowercase
    text = text.lower()
    # Tokenization (you can use more advanced tokenization techniques)
    tokens = text.split()
    # Remove punctuation and stopwords (you can enhance this step)
    tokens = [word.strip('.,!?()[]{}"\'') for word in tokens if word.isalpha()]
    #print(tokens)
    # Join tokens back to text
    return " ".join(tokens)

df['Statement'] = df['Statement'].apply(preprocess_text)

print(df['Statement'][0])
```

###**Train-Test Split**

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['Statement'], df['Label'], test_size=0.2, random_state=42)
```

###**TF-IDF Vectorization**

```python
# TF-IDF feature extraction
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

###**Word2Vec Embeddings**

```python
# First, we need to tokenize the sentences for Word2Vec training
tokenized_statements = [sentence.split() for sentence in df['Statement']]
word2vec_model = Word2Vec(tokenized_statements, vector_size=100, window=5, min_count=1, sg=0)
```

```python
def get_word2vec_features(text, model):
    words = text.split()
    # Remove words that are not in the vocabulary
    words = [word for word in words if word in model.wv]
    if len(words) == 0:
        # If no words are in the vocabulary, return a vector of zeros
        return np.zeros(model.vector_size)
    # Calculate the mean vector of word embeddings
    word_vectors = [model.wv[word] for word in words]
    return np.mean(word_vectors, axis=0)

X_train_word2vec = np.array([get_word2vec_features(text, word2vec_model) for text in X_train])
X_test_word2vec = np.array([get_word2vec_features(text, word2vec_model) for text in X_test])
```

###**Machine Learning Model Declaration**

```python
models = {
    "Logistic Regression": LogisticRegression(verbose=2),
    "Random Forest": RandomForestClassifier(verbose=2),
    "SVM": SVC(verbose=2),
}
```

###**Deep Learning Model (GPT-4)**

```python
gpt4_classifier = pipeline("text-classification", model="EleutherAI/gpt-neo-2.7B")
```

```python
# Initialize results dictionary to store accuracies
results = {"Model": [], "TF-IDF Accuracy": [], "Word2Vec Accuracy": []}
```

##**Training and testing ML models**

```python
# Training and testing loop
for model_name, model in models.items():
    # Train and test with TF-IDF
    model.fit(X_train_tfidf, y_train)
    y_pred_tfidf = model.predict(X_test_tfidf)
    accuracy_tfidf = accuracy_score(y_test, y_pred_tfidf)

    # Train and test with Word2Vec
    model.fit(X_train_word2vec, y_train)
    y_pred_word2vec = model.predict(X_test_word2vec)
    accuracy_word2vec = accuracy_score(y_test, y_pred_word2vec)

    # Store results
    results["Model"].append(model_name)
    results["TF-IDF Accuracy"].append(accuracy_tfidf)
    results["Word2Vec Accuracy"].append(accuracy_word2vec)
```

###**GPT 3**
